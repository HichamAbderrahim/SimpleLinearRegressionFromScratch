We notice that we need to calculate the mean more than once, so I write a function for the shortcut

```python
def calc_mean(table):
  """ function for calculating the mean """
  sum = 0
  for i in range(len(table)):
    sum += table[i]
  return sum / len(table) # Or just return sum(table) / len(table)
```

I prepare two tables that represent the coordinates of the points

```python
X = [2, 4, 1, 2, 5]
Y = [1, 3, 5, 3, 5]
```

I extract the number of points by taking the smallest possible length between the two tables

```python
n = len(X) if len(X) < len(Y) else len(Y) # Or n = min(len(Xi), len(Yi))
if n < 1:
  raise Exception("Error: The passed value cannot be less than 1")
```

Now we use the function that I wrote to calculate the mean of the Y values because we need it to calculate the numerator

```python
y_mean = calc_mean(Y)
```

I calculate the numerator

```python
numerator = 0
for i in range(n):
  numerator += X[i]*(Y[i] - y_mean)
```

I use the function again to calculate the mean of the X values because this time we need it to calculate the denominator

```python
x_mean = calc_mean(X)
```

I calculate the denominator



```python
denominator = 0
for i in range(n):
  denominator += X[i]*(X[i] - x_mean)
```

Now that we have calculated the numerator and denominator, I do the division process to get the value of a (the slope)

```python
a = numerator / denominator
```

We already have the mean values of x and y and we also calculated the value of a (the slope) so I apply the formula to get b (the point of intersection with the y axis)

```python
b = y_mean - a * x_mean
```

Now I print the values of a and b

```python
print("a = ", round(a, 3))
print("b = ", round(b, 3))             
```
